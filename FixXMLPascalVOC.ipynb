```python
#https://www.guru99.com/manipulating-xml-with-python.html
#https://docs.python.org/3/library/xml.etree.elementtree.html#module-xml.etree.ElementTree
import os
import xml.etree.ElementTree as XML
import cv2
```

```python
#open xml
counter = 0
directory = "D:/DATASETS/VANS/take_3_skateboard_maindor_take_1"

def castBoundingBoxesToInt(xml_files):
    global counter
    for xml_file in xml_files:
        tree = XML.parse(xml_file)
        root = tree.getroot()
        for obj in root.findall('object'):
            for boundingbox in obj.findall('bndbox'):
                for item in boundingbox:
                    if item.text:
                        new_value =  int(float(item.text))
                        item.text = str(new_value)
        counter = counter+1
        print(counter)
        filename = os.path.basename(xml_file)
        newpath = os.path.join(directory,filename)        
        tree.write(newpath)
        
def drawBoundingBoxes(xml_files):
    global counter
    for xml_file in xml_files:
        counter = counter+1
        
        if counter % 50 == 0:
        
            filename = os.path.basename(xml_file)
            jpgfilename = os.path.splitext(filename)[0]+'.jpg'
            print(jpgfilename)
            tree = XML.parse(xml_file)
            root = tree.getroot()
            for obj in root.findall('object'):
                for boundingbox in obj.findall('bndbox'):
                    if boundingbox.find('xmin').text:
                        xmin = int(float(boundingbox.find('xmin').text))
                        ymin =  int(float(boundingbox.find('ymin').text))
                        xmax =  int(float(boundingbox.find('xmax').text))
                        ymax =  int(float(boundingbox.find('ymax').text))
                        img = cv2.imread(os.path.join(directory,jpgfilename))
                        cv2.rectangle(img, (int(xmin),int(ymin)),(int(xmax),int(ymax)), (255,0,0), 2)
                        cv2.imshow("result", img)
                        cv2.waitKey(0)
                        cv2.destroyAllWindows()
            
def return_xmls(folder):
    global counter
    counter = 0
    all_xmls = []
    files = os.listdir(folder)
    for f in files:
        if os.path.splitext(f)[1] == ".xml":
            newpath = os.path.join(directory,f)
            all_xmls.append(newpath)
    return all_xmls
            
xml_files = return_xmls(directory)
castBoundingBoxesToInt(xml_files)
#drawBoundingBoxes(xml_files)

            
print("done")
```
