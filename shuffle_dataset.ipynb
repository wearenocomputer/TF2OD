```python
import os
import cv2
import uuid
import shutil
import xml.etree.ElementTree as XML
directory = "C:/Users/paperspace/Downloads/take_3/all"


path_shuffled = os.path.join(directory, "shuffled")
if not os.path.exists(path_shuffled):
    os.mkdir(path_shuffled)

    
def doShuffle(files):
    for xml_file in files:
       
        filename = os.path.basename(xml_file)
        jpgfilename = os.path.splitext(filename)[0]+'.jpg'
        xmlfilename = os.path.splitext(filename)[0]+'.xml'
        
        
        origin_xml_full_path = xml_file
        origin_jpg_full_path = os.path.join(directory, jpgfilename)
        
        _uuid = str(uuid.uuid4())
        
        #change the xml too
        tree = XML.parse(xml_file)
        root = tree.getroot()
        
       
        for obj in root.findall('filename'):
            obj.text = str(_uuid+".jpg")
        
        new_xml_full_path =  os.path.join(path_shuffled, _uuid+".xml")
        new_jpg_full_path =  os.path.join(path_shuffled, _uuid+".jpg")
        
        
        tree.write(new_xml_full_path)
        #print(new_xml_full_path)
        print(new_jpg_full_path)
        #shutil.copy(origin_xml_full_path, new_xml_full_path)
        shutil.copy(origin_jpg_full_path, new_jpg_full_path)
        

def return_xmls(folder):
    all_xmls = []
    files = os.listdir(folder)
    for f in files:
        if os.path.splitext(f)[1] == ".xml":
            newpath = os.path.join(directory,f)
            all_xmls.append(newpath)
    return all_xmls

xml_files = return_xmls(directory)
doShuffle(xml_files)

```
